## Description
Triangular Nim is a two-player strategy game played on a pyramid of matchsticks.
The board has 5 rows arranged in a triangle:

```
|    <- Row 1: 1 stick
|  |    <- Row 2: 2 sticks
|  |  |    <- Row 3: 3 sticks
|  |  |  |    <- Row 4: 4 sticks
|  |  |  |  |    <- Row 5: 5 sticks
```

This version starts with 15 matchsticks and uses the Search Tree algorithm so the AI plays the best moves.

### Rules Summary:
- On each turn, a player picks **one row** and removes **at least 1** stick from it
- You can remove **as many sticks as you like** from that row (up to all of them)
- The player who takes the **very last stick loses**

## CurrentBoard

The board state is represented as an array of numbers one value per row,
each value being the number of sticks remaining in that row.
For example `(1, 2, 3, 4, 5)` is the starting position and `(0, 0, 0, 0, 0)` is game over.

The `state_of_board()` method returns:
- "U" -> Game unfinished (at least one stick remains)
- "LAST_TAKEN" -> All sticks gone (game over)

````python
class CurrentBoard:

    board = ()

    def __init__(self, string_def=INITIAL_ROWS):
        self.board = string_def # Every move = new CurrentBoard
        self.state = self.state_of_board() # Check the state of the board

    def display(self, game_display=False):
        for row_idx in range(len(self.board)):
            sticks = self.board[row_idx] # Num of sticks left in this row
            max_sticks = INITIAL_ROWS[row_idx] # Num of sticks started in this row
            stick_str = "  ".join("|" for i in range(sticks))
            empty_str = "  ".join(" " for i in range(max_sticks - sticks))
            if sticks > 0 and max_sticks - sticks > 0:
                row_str = stick_str + "     " + empty_str
            else:
                row_str = stick_str if sticks > 0 else empty_str
            print(f"{row_str}   <- Row {row_idx + 1} ({sticks} left)")

    def other(self, piece):
        if piece == "X":
            return "O"
        return "X"

    def state_of_board(self):
        if sum(self.board) == 0:
            return "LAST_TAKEN" # Game over
        return "U" # Unfinished game

    def all_possible_moves(self, player_piece):
        possible_moves = []
        for row_idx in range(len(self.board)):
            sticks_in_row = self.board[row_idx]
            if sticks_in_row == 0:
                continue # Skip empty rows
            for take in range(1, sticks_in_row + 1):
                new_board = list(self.board)
                new_board[row_idx] -= take # Remove sticks from chosen row
                possible_moves.append(CurrentBoard(string_def=tuple(new_board)))
        return possible_moves


    def evaluation(self):
        total_sticks = sum(self.board)
        if total_sticks <= 1:
            return -0.5 # Only 1 stick left -> next player must take it and lose
        elif total_sticks <= 4:
            return 0.5 # Few sticks left
        else:
            return 0.0 # Many sticks left -> Neutral
````

The evaluation function estimates the board value based on the total number of sticks remaining. If only 1 stick is left the position is considered losing since the next player is forced to take it. If between 2 and 4 sticks remain the position is considered winning since the opponent is close to being forced into taking the last stick. Otherwise the position is considered neutral as there are too many sticks remaining to make a judgement.

## SearchTreeNode

Builds the full Minimax game tree.

- **Even ply** (ply % 2 == 0) -> AI's turn -> **Maximise** (pick child with highest value)
- **Odd ply**  (ply % 2 == 1) -> Player's turn -> **Minimise** (pick child with lowest value)

Leaf node values (LAST_TAKEN means the board is empty):
- ply % 2 == 1 -> AI just took last stick -> AI loses -> value = -1
- ply % 2 == 0 -> Human just took last stick -> AI wins -> value = 1

````python
class SearchTreeNode:

    def __init__(self, board_instance, playing_as, ply=0):
        self.children = []
        self.value_is_assigned = False
        self.ply_depth = ply
        self.current_board = board_instance
        self.move_for = playing_as

        if self.current_board.state == "U":
            self.generate_children() # Generate all possible moves
        else:
            # ply % 2 == 0: computer took last stick -> computer loses -> value = -1
            # ply % 2 == 1: human took last stick -> human loses -> value = 1
            if (self.ply_depth % 2) == 0: # Even ply
                self.value = -1 # AI took last stick, AI loses
            else:
                self.value = 1 # Human took last stick, AI wins
            self.value_is_assigned = True

    def min_max_value(self):
        if self.value_is_assigned:
            return self.value

        self.children = sorted(self.children, key=lambda x: x.min_max_value())

        if (self.ply_depth % 2) == 0:
            self.value = self.children[-1].value  # Max
        else:
            self.value = self.children[0].value   # Min

        self.value_is_assigned = True
        return self.value

    def generate_children(self):
        for board_for_next_move in self.current_board.all_possible_moves(self.move_for):
            self.children.append(
                SearchTreeNode(
                    board_for_next_move, # New board after this move
                    self.current_board.other(self.move_for), # Switch to other player
                    ply=self.ply_depth + 1 # Go one level deeper
                )
            )
````

### Why it crashed

From the starting position (1,2,3,4,5) there are 15 possible first moves.
Each of those branches into further moves recursively until all sticks are gone, creating a tree with thousands of nodes loaded into memory at once.
Unlike TicTacToe which has at most 9 moves, Triangular Nim has up to 15 sticks meaning the tree grows much faster and exhausts available RAM.

### Proposed Fix

Introduce a MAX_PLY depth limit in SearchTreeNodeV2. Instead of exploring all the way to the end of the game, the tree stops at depth 6 and calls an evaluation function to estimate the value of that position. During actual gameplay this is not
a problem because the tree is rebuilt per move from the current board, which is already reduced, keeping memory usage manageable.

## Testing and Evaluation: Update SearchTreeNode

````python
# Misère Nim evaluation function
# Uses proven mathematical theory to correctly score any board position
def misere_evaluation(board):
    """
    Returns True if the player about to move WINS from this position.

    Misère Nim theory:
    - If all rows have 0 or 1 sticks: even number of single sticks = win
    - Otherwise: nim-sum (XOR of all rows) != 0 = win
    """
    all_small = all(x <= 1 for x in board)
    if all_small:
        active = sum(1 for x in board if x == 1)
        return (active % 2 == 0) and (active > 0)
    else:
        nim = 0
        for x in board:
            nim ^= x
        return nim != 0
````

### Misere Nim Evaluation Function

The evaluation function is based on proven mathematical theory for Misere Nim (the version where taking the last stick loses). It correctly identifies whether the player about to move is in a winning or losing position:

- If all remaining rows have 0 or 1 sticks (endgame): an even number of
  single-stick rows is a winning position because you can always leave an odd number for your opponent.
- If at least one row has more than 1 stick (mid-game): the nim-sum
  (binary XOR of all row values) being non-zero means the current player can win by moving to a nim-sum of zero for the opponent.

This is far more accurate than simply counting total sticks.

````python
MAX_PLY = 6

class SearchTreeNodeV2:

    def __init__(self, board_instance, playing_as, ply=0):
        self.children = []
        self.value_is_assigned = False
        self.ply_depth = ply
        self.current_board = board_instance
        self.move_for = playing_as

        if self.current_board.state == "U" and self.ply_depth < MAX_PLY:
            self.generate_children()
        else:
            if self.current_board.state == "LAST_TAKEN":
                # whoever just moved took the last stick and loses
                # odd ply = AI just moved = AI loses = -1
                # even ply = human just moved = AI wins = +1
                if (self.ply_depth % 2) == 1:
                    self.value = -1  # AI took last stick, AI loses
                else:
                    self.value = 1   # Human took last stick, AI wins
            else:
                # depth cutoff, use Misère evaluation
                current_can_win = misere_evaluation(self.current_board.board)
                if (self.ply_depth % 2) == 0:
                    # AI is about to move: winning position is good for AI
                    self.value = 0.5 if current_can_win else -0.5
                else:
                    # Human is about to move: human winning is bad for AI
                    self.value = -0.5 if current_can_win else 0.5
            self.value_is_assigned = True

    def min_max_value(self):
        if self.value_is_assigned:
            return self.value
        self.children = sorted(self.children, key=lambda x: x.min_max_value())
        if (self.ply_depth % 2) == 0:
            self.value = self.children[-1].value  # AI maximises
        else:
            self.value = self.children[0].value   # Human minimises
        self.value_is_assigned = True
        return self.value

    def generate_children(self):
        for board_for_next_move in self.current_board.all_possible_moves(self.move_for):
            self.children.append(
                SearchTreeNodeV2(
                    board_for_next_move,
                    self.current_board.other(self.move_for),
                    ply=self.ply_depth + 1
                )
            )
````

## Game Play

````python
def play_Nim():
    response = input("Do you wish to play first (y/n)? ")
    players_turn = (response == "y" or response == "Y")

    cb = CurrentBoard()
    player_is_playing = "X"  # Human is X, AI is O

    print()
    cb.display()
    print()

    while True:

        # Human turn
        if players_turn:
            while True:
                try:
                    row = int(input(f"Your turn. Which row? (1-{len(cb.board)}): ")) - 1
                    if 0 <= row < len(cb.board) and cb.board[row] > 0:
                        break
                    print("Invalid row or that row is empty. Try again.")
                except ValueError:
                    print("Please enter a valid number.")

            while True:
                try:
                    take = int(input(f"How many from Row {row + 1}? (1-{cb.board[row]}): "))
                    if 1 <= take <= cb.board[row]:
                        break
                    print(f"Please enter a number between 1 and {cb.board[row]}.")
                except ValueError:
                    print("Please enter a valid number.")

            new_board = list(cb.board)
            new_board[row] -= take
            cb = CurrentBoard(string_def=tuple(new_board))
            print(f"\nYou took {take} from Row {row + 1}.")
            print()
            cb.display()
            print()

        # AI turn
        else:
            print("\nAI is making the move, playing as O")
            print("Constructing the Search Tree...")
            search_tree = SearchTreeNodeV2(cb, cb.other(player_is_playing))

            print("Assigning values to the nodes...Applying the Min Max algorithm")
            search_tree.min_max_value()

            print("Choosing the best move")
            best_board = search_tree.children[-1].current_board

            ai_row  = next(i for i in range(len(cb.board)) if cb.board[i] != best_board.board[i])
            ai_take = cb.board[ai_row] - best_board.board[ai_row]
            cb = best_board

            print(f"\nAI took {ai_take} from Row {ai_row + 1}.")
            print()
            cb.display()
            print()

        # Check end state
        if cb.state != "U":
            print()
            if players_turn:
                print("You took the last stick, you lose!")
            else:
                print("AI took the last stick, you win!")
            break

        players_turn = not players_turn

    print()
    again = input("Play again? (y/n): ").strip().lower()
    if again == 'y':
        play_Nim()
    else:
        print("Thanks for playing!")
````
